### Class Sizes  
School K-8 Average Class Size for 2017-2018 school year

```python
df_clsz = pd.read_excel('./data/February2018_Avg_ClassSize_School.xlsx')
df_clsz.head(10)
```

```python
df_clsz.rename(columns=(df_clsz.loc[6]), inplace = True)
df_clsz = df_clsz[7:]
df_clsz.head()
```

```python
df_clsz['BN'] = [x[2:6] for x in df_clsz.DBN]
```

```python
df_clsz.groupby(['Grade Level', 'Program Type'])
```

** * creating variable for program type and each grade* **

```python
cls_sz = {}
```

```python
[yr for yr in df_clsz['Grade Level'].unique()]
```

```python
[yr for yr in df_clsz['Program Type'].unique()]
```

```python
for yr in df_clsz['Grade Level'].unique():
    for typ in df_clsz['Program Type'].unique():
        name = str(yr) +"_"+ typ
        cls_sz[name] = df_clsz[(df_clsz['Grade Level'] == yr) & (df_clsz['Program Type'] == typ)]
```

```python
name
```

```python
cls_sz[name][['Number of Students', 'Number of Classes', 'Average Class Size']]
```

```python
for name in cls_sz.keys():
    nu_c = [col+'_'+name for col in cls_sz[name]
            [['Number of Students', 'Number of Classes', 'Average Class Size']]]
    cls_sz[name].rename(columns=dict(zip(cls_sz[name]
                                         [['Number of Students', 'Number of Classes', 'Average Class Size']],
                                         nu_c)), inplace=True) 
```

```python
df_clssz_nu = pd.concat(cls_sz)
df_clssz_nu.reset_index(inplace=True)
df_clssz_nu.drop(['level_0', 'level_1'], axis=1, inplace = True)
```

```python
list(df_clssz_nu)
```

```python
df_clssz_nu.to_csv('./data/Class_sizes_reshaped_on_dbn.csv')
```
