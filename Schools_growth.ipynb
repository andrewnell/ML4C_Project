**Goal: ** Identify schools that have helped students grow.

- Create a measure for growth: has a school taken low-performing 9th graders and helped them achieve?

quantile methods:
- [HPGG08PROF] look at incoming 8th grade scores for  ELA and Math for 8th grade exam, and put schools into quantiles in year x
- do the same for 4 year diploma rate four years later [HPGPERDIP4AVG]
- see which schools do better (move into a higher quantile)

Growth score (based on DOE methods)
- for all schools with similar entering proficiency, compare their graduation rates.  The score is the percent of schools that it does better than.

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import pylab as pl
from scipy import stats
%pylab inline
```

```python
df = pd.read_csv('Schmelt_bg.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

## Quantile Method - high schools
- only have average 8th grade proficiency for incoming 9th graders from years 2008 to 2012
- will take only years 2012 to 2016 for 4 yr grad rate
- will subtract the difference in quantiles to get a score

*** quantiles (dec) for avg 8th grade providiency ***

```python
#drop columns that don't have any values
df_8prof = df[df.variable == 'HPGG08PROF'].dropna(axis=1, how='all')
```

```python
#drop school that don't have any entries for any year
df_8prof = df_8prof[df_8prof.variable == 'HPGG08PROF'].dropna(axis=0, how='all', subset=['2008', '2009', '2010',
       '2011', '2012'])
```

```python
for yr in ['2008', '2009', '2010','2011', '2012']:
    df_8prof[yr] = pd.to_numeric(df_8prof[yr])
    df_8prof[('prof8dec_'+yr)] = pd.qcut(df_8prof[yr], 10, labels=False)
```

```python
df_8prof.head()
```

*** quantiles (dec) for 4 year graduation rates ***
- will need to get these from data we already have

```python
df_grdrt = df[df.variable == 'HPGPERDIP4AVG'].dropna(axis=1, how='all')
```

```python
df_grd = pd.read_csv('./data/2017_grad_rates_reshaped.csv')
```

Possible features:  
(could take august or 5, 6 year grad rates, too)

- 'Total Grads_% of cohort_2008_4 year June'

```python
cols = ['DBN', 'School Name','BN']
for yr in ['2007', '2008', '2009','2010', '2011']:
    cols.extend(['Total Grads_% of cohort_{}_4 year June'.format(yr)])
print (cols)
```

```python
df_grd = df_grd[cols]
```

```python
for yr in ['2008', '2009', '2010','2011', '2012']:
    df_grd['Total Grads_% of cohort_{}_4 year June'.format(str(int(yr)-1))] = \
        pd.to_numeric(df_grd['Total Grads_% of cohort_{}_4 year June'.format(str(int(yr)-1))], errors='coerce')
    df_grd[('grddec_'+yr)] = pd.qcut(df_grd['Total Grads_% of cohort_{}_4 year June'.format(str(int(yr)-1))], 10, labels=False)
```

```python
df_grd.head()
```

*** merge grad df and 8g prof df, and take the difference between 8th grade and grad rates as the growth score ***

```python
print ('Rows in df_grd: ', len(df_grd))
print ('Unique schools in df_grd: ', len(df_grd.BN.unique()))
print ('Rows in df_8prof: ', len(df_8prof))
print ('Unique schools in df_8prof: ', len(df_8prof.BNLONG.unique()))
df_ = pd.merge(df_8prof, df_grd, left_on='BNLONG', right_on='BN', how='inner')
print ('Rows in new merged df: ', len(df_))
print ('Unique schools in new merged df: ', len(df_.BNLONG.unique()))
```

No duplicated values in either dataframe. Lose 111 from df_grd, and 4 from df_8prof.

```python
df_.head()
```

```python
for yr in ['2008', '2009', '2010','2011', '2012']:
    df_['HS_growth_score'+str(int(yr)+3)] = df_['grddec_'+yr] - df_['prof8dec_'+yr]
```

```python
df_.head()
```

- prof8dec is the decile that the school falls into, based on the average 8th grade proficiency of the incoming 9th grade class (data from SCHMA) (year corresponds to the year that the 8th graders were in the 9th grade)
- grddec is the decile that the school falls into, based on the average 4 yr June graduation rate (year corresponds to the year the graduating class was in the 9th grade) ( data from 2017_grad_rates_reshaped.csv)
- HSgrowthscore is the difference between the grad decile and the 8th grade proficiency decile. the year corresponds to the cohort year, ie. the year the students were in 9th grade. This score represents movement in proficiency for the same group of students.
